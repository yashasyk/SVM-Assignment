```python
# SVM Classification
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler

from sklearn.svm import SVC
from sklearn.model_selection import GridSearchCV #automated hyperparameter tuning
from sklearn.metrics import classification_report

from sklearn.metrics import accuracy_score, confusion_matrix
from sklearn.model_selection import train_test_split, cross_val_score
```

```python
salary_train = pd.read_csv("C:\\My Computer\\Documents\\Excler\\Assignment\\SVM\\SalaryData_Train(1).csv")
salary_test = pd.read_csv("C:\\My Computer\\Documents\\Excler\\Assignment\\SVM\\SalaryData_Test(1).csv")
```

```python
category = ['workclass', 'education','maritalstatus','occupation','relationship','race','sex','native']
```

```python
from sklearn.preprocessing import LabelEncoder

labelencoder = LabelEncoder()
for i in category:
    salary_train[i] = labelencoder.fit_transform(salary_train[i])
    salary_test[i] = labelencoder.fit_transform(salary_test[i])
```

```python
salary_train.head(5)
```

```python
X_train = salary_train.iloc[:,0:13]
X_test = salary_test.iloc[:,0:13]
y_train = salary_train.iloc[:,13]
y_test = salary_test.iloc[:,13]
```

```python
def norm_func(i):
    x = (i-i.min())/(i.max()-i.min())
    return (x)
X_train = norm_func(X_train)
X_test =  norm_func(X_test)
```

```python
X_train
```

### Grid Search CV

```python
clf = SVC(kernel = 'poly')
param_grid = [{'gamma':[0.2,0.1,0.01],'C':[0,0.1,0.001,0.0001] }] 
gsv = GridSearchCV(clf,param_grid,cv=10) #cv=cross validation
gsv.fit(X_train,y_train)
```

```python
gsv.best_params_, gsv.best_score_
```

```python
clf = SVC(C= 0.1, gamma = 0.2) #input the C and gamma values from the above cell(best params and score)
clf.fit(X_train , y_train)
y_pred = clf.predict(X_test)
acc = accuracy_score(y_test, y_pred) * 100
print("Accuracy =", acc)
confusion_matrix(y_test, y_pred)
```
